```python
from netCDF4 import Dataset
Full3D = Dataset('/glade/p/work/mlevy/G_term_comparisons/C_write_G.pop.h.nday1.0001-01-04-03456.nc')
Diff_3D1D = Dataset('/glade/p/work/mlevy/G_term_comparisons/C_minus_C1D.pop.h.nday1.0001-01-04-03456.nc')

tlat = Full3D.variables['TLAT']
tlon = Full3D.variables['TLONG']
ulat = Full3D.variables['ULAT']
ulon = Full3D.variables['ULONG']
```

```python
data = dict()
data['Full3D'] = dict()
data['3D-1D'] = dict()

import numpy as np

for fieldname in ('GT', 'GS', 'GU', 'GV'):
    temp = np.array(Full3D.variables[fieldname])
    temp[temp > 1e30] = np.nan
    data['Full3D'][fieldname] = temp
    
for fieldname in ('TEND_TEMP', 'TEND_SALT', 'TEND_U', 'TEND_V'):
    temp = np.array(Diff_3D1D.variables[fieldname])
    temp[temp > 1e30] = np.nan
    data['3D-1D'][fieldname] = temp
```

```python
# Print min and max for all values we are comparing
for runtype, runval in sorted(data.items()):
    for varname, varval in sorted(data[runtype].items()):
        print varname, ':', np.nanmin(data[runtype][varname]), np.nanmax(data[runtype][varname])
```
